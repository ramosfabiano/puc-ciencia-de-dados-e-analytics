## 4. Pré-Processamento de dados


### 4.1. Tratamento de Missings e Limpeza


Iniciamos então a etapa de pré-processamento pelo tratamento de missings.

```python
#ms.matrix(listings,figsize=(15, 5))
print(listings.isnull().sum())
```

Observamos que os seguintes atributos necessitam de tratamento de *missings*:

* **host_is_superhost**:  atribuiremos o valor *False*.
* **host_listings_count**:  atribuiremos o valor *0.0*.
* **bathrooms**: eliminaremos a coluna, pois são *todos* nulos.
* **bedrooms** : eliminaremos o registro, pois é um atributo demaisado relevante para ser estimado.
* **beds** : atribuiremos a mediana da coluna.
* **review_scores_*** : atribuiremos a mediana da coluna.



```python
# dataset original sem missings para separacao conjunto testes
listings_no_missings = listings.notnull()

# preserva dataset original
listings_clean = listings

# host_is_superhost: atribuiremos o valor False.
listings_clean['host_is_superhost'].fillna('f', inplace=True)
listings_clean['host_is_superhost'] = listings_clean['host_is_superhost'].map({'t': True, 'f': False})

# host_listings_count:  atribuiremos o valor 0.0.
listings_clean['host_listings_count'].fillna(0.0, inplace=True)

# bathrooms: eliminaremos a coluna, pois são todos nulos.
listings_clean = listings_clean.drop(columns=['bathrooms'])

# bedrooms : eliminaremos o registro, pois é um atributo chave.
listings_clean = listings_clean.dropna(axis=0, subset=['bedrooms'])

# beds : atribuiremos a mediana da coluna.
listings_clean['beds'].fillna(listings_clean['beds'].median(), inplace=True)

# review_scores_* :  eliminaremos o registro, pois é um atributo chave.
#listings_clean = listings_clean.dropna(axis=0, subset=['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin' , 'review_scores_communication', 'review_scores_location', 'review_scores_value'])

# review_scores_* :  atribuiremos a mediana da coluna.
for c in ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin' , 'review_scores_communication', 'review_scores_location', 'review_scores_value']:
  listings_clean[c].fillna(listings_clean[c].median(), inplace=True)

# exibe novo dataset 'limpo'
listings_clean.info()
listings_clean.head().transpose()
```

Revisitamos agora as oportunidades de pré-processamento identificadas durante a etapa enterior:

* Eliminação da coluna **bathroooms**: já realizada anteriormente durante o tratamento de *missings*.
* Conversão do campo **host_listings_count** de *float* para *int*.
* Tratamento da coluna **price**, transformando em numérico.


```python
# host_list_count: float => int
listings_clean['host_listings_count'] = listings_clean['host_listings_count'].astype(int)
 
# price: remove ['$' , ','] e converte para numérico
listings_clean['price'] = listings_clean['price'].str.replace('[\$,]', '', regex=True).astype(float)

listings_clean.info()
listings_clean.head(5).transpose()
```

### 4.2. Engenharia de Atributos

Também identificamos durante a análise exploratória preliminar a necessidade de tratamento da coluna **amenities**. 

O tratamento deste atributo é complexo, pois é um campo não estruturado. Por outro lado, entendemos que é um atributo chave na avaliação do preço final para locação, portanto vamos tentar oferecer algum tratamento.

Primeiramente buscamos uma lista das diferentes comodidades:


```python
# dicionario de amenidades individuais
unique_amenities_dict = {}

df = listings_clean
for idx, row in df.iterrows():
  # o valor do atributo é uma string com a list de amenidades 
  # ex: '["Wifi", "Dedicated workspace", "Kitchen"]' 
  amenities_str = df.loc[idx,'amenities']
  # removendo caracteres especiais e espaços multiplos
  amenities_str = re.sub(r'[\[\]\}\{\-\$\(\)}\"]','', amenities_str)
  amenities_str = re.sub(' +', ' ', amenities_str)
  amenities_str = re.sub(', ', ',', amenities_str)
  amenities_str = re.sub(' ', '_', amenities_str)
  amenities_str = amenities_str.lower()
  # atualiza o dado 'limpo', para acelerar a segunda passagem
  df.loc[idx,'amenities'] = amenities_str
  # divisao em comodidades inviduais, como uma lista 
  amenities_list = amenities_str.split(',')
  # contabilizando no dicionário 
  for individual_amenity in amenities_list:
    if individual_amenity in unique_amenities_dict:
      unique_amenities_dict[individual_amenity] += 1
    else:
      unique_amenities_dict[individual_amenity] = 1

print(len(unique_amenities_dict))
```

Após o tratamento inicial, podemos observar 3498 comodidades únicas. 

Nosso plano é codificar as diferentes comodidades em diferentes atributos, mas este número ainda é demasiado elevado. Idealmente esta lista deveria ser inspecionada por um especialista de domínio no sentido de apontar as comodidades mais significativas. Para efeito deste trabalho, vamos usar a heurística de selecionar as 20 comodidades mais frequentes.



```python
top_20_amenities = dict ((sorted(unique_amenities_dict.items(), key=lambda x:x[1], reverse=True))[0:20])
for k in top_20_amenities.keys():
  print(k)
```

Examinamos então o histograma das comodidades individiduais selecionadas.

```python
plt.figure(figsize=(5, 4))
plt.bar(list(top_20_amenities.keys()), top_20_amenities.values())
plt.xticks('')
plt.xlabel('Top 20 Comodidades')
plt.ylabel('Ocorrências')
plt.show()
```

A seguir, criamos os atributos referentes a cada uma das comodidades selecionadas.

```python
# preservando o dataset antes da inclusao
listings_clean_expanded = listings_clean

# criando uma coluna para cara comodidade, iniciando como False
for key in top_20_amenities.keys():
  listings_clean_expanded[key] = False

# repetimos o processamento anterior, mas desta vez
# preenchendo os valores dos novos atributos
# note que os dados já foram 'limpos' na primeira passada
df = listings_clean_expanded
for idx, row in df.iterrows():
  amenities_str = df.loc[idx,'amenities']
  amenities_list = amenities_str.split(',')  
  for individual_amenity in amenities_list:
    if individual_amenity in df.columns:
      df.loc[idx, individual_amenity] = True

# removendo a coluna original, não mais necessária
listings_clean_expanded.drop(columns=['amenities'])

listings_clean_expanded.info()
```
