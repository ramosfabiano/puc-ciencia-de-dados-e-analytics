## Preparação dos dados


### Pré-Processamento

Iniciamos então o pré-processamento dos dados pelo tratamento de *missings*:

```python
# preserva dataset original
listings_clean = listings

# host_is_superhost
listings_clean['host_is_superhost'].fillna('f', inplace=True)
listings_clean['host_is_superhost'] = listings_clean['host_is_superhost'].map({'t': True, 'f': False})

# bathroom
listings_clean = listings_clean.drop(columns=['bathrooms'])

# bedrooms
listings_clean = listings_clean.dropna(axis=0, subset=['bedrooms'])

# beds
listings_clean['beds'].fillna(np.floor(listings_clean['beds'].median()), inplace=True)

# review_scores_rating
listings_clean['review_scores_rating'].fillna(listings_clean['review_scores_rating'].median(), inplace=True)
```

Realizamos também outras atividades de pré-processamento:

* Conversão do campo **host_is_superhost** de *float* para *int*.
* Tratamento da coluna **price**, transformando em numérico.
* Codificação dos atributos **neighbourhood_cleansed**, **room_type** e **property_type** como atributos númericos.


```python
# host_list_count: float => int
listings_clean['host_is_superhost'] = listings_clean['host_is_superhost'].astype(int)

# price: remove ['$' , ','] e converte para numérico
listings_clean['price'] = listings_clean['price'].str.replace('[\$,]', '', regex=True).astype(float)
listings_clean['price'] = np.log(listings_clean['price'])

# codificação dos atributos
le = LabelEncoder()
le.fit(listings_clean['neighbourhood_cleansed'].values)
listings_clean['neighbourhood_cleansed'] =  le.transform(listings_clean['neighbourhood_cleansed'].values)
le.fit(listings_clean['room_type'].values)
listings_clean['room_type'] =  le.transform(listings_clean['room_type'].values)
le.fit(listings_clean['property_type'].values)
listings_clean['property_type'] = le.transform(listings_clean['property_type'].values)
```

O tratamento do atributo **amenities**  se mostrou complexo, pois é um campo não estruturado.

Entendemos que este é um atributo chave na avaliação do preço final para locação, portanto tentamos oferecer um tratamento adequado.

Primeiramente listamos as diferentes comodidades listadas:


```python
# dicionario de amenidades individuais
unique_amenities_dict = {}

df = listings_clean
for idx, row in df.iterrows():
  # o valor do atributo é uma string com a list de amenidades
  # ex: '["Wifi", "Dedicated workspace", "Kitchen"]'
  amenities_str = df.loc[idx,'amenities']
  # removendo caracteres especiais e espaços multiplos
  amenities_str = re.sub(r'[\[\]\}\{\-\$\(\)}\"]','', amenities_str)
  amenities_str = re.sub(' +', ' ', amenities_str)
  amenities_str = re.sub(', ', ',', amenities_str)
  amenities_str = re.sub(' ', '_', amenities_str)
  amenities_str = amenities_str.lower()
  # atualiza o dado 'limpo', para acelerar a segunda passagem
  df.loc[idx,'amenities'] = amenities_str
  # divisao em comodidades inviduais, como uma lista
  amenities_list = amenities_str.split(',')
  # contabilizando no dicionário
  for individual_amenity in amenities_list:
    if individual_amenity in unique_amenities_dict:
      unique_amenities_dict[individual_amenity] += 1
    else:
      unique_amenities_dict[individual_amenity] = 1

print(f'Comodidades únicas: {len(unique_amenities_dict)}')
```

A seguir, criamos novos atributos referentes à cada uma das comodidades, no estilo *one-hot-encoding*. Mantivemos apenas as 20 comodidades mais frequentemente listadas.



```python
# top 20
top_20_amenities = dict ((sorted(unique_amenities_dict.items(), key=lambda x:x[1], reverse=True))[0:20])

print(f'Número inicial de atributos: ', len(listings_clean.columns))

# criando uma coluna para cada comodidade
for key in top_20_amenities.keys():
  df.insert(0, key, 0.0)

# preenchendo os valores dos novos atributos
# note que os dados já foram 'limpos' na primeira passada
df = listings_clean
for idx, row in df.iterrows():
  amenities_str = df.loc[idx,'amenities']
  amenities_list = amenities_str.split(',')
  for individual_amenity in amenities_list:
    if individual_amenity in df.columns:
      df.loc[idx, individual_amenity] = 1.0

print(f'Novo número de atributos: ', len(listings_clean.columns))

# descartando a coluna original, não mais necessária
listings_clean = listings_clean.drop(columns=['amenities'])
```

### Seleção de Atributos

Com o objetivo de potencialmente melhorar os resultados e interpretabilidade dos modelos, realizamos também uma etapa de *feature selection*. Focamos em apenas uma metodologia: **importância de atributos**, apoiada no uso de *ExtraTrees*.





```python
X = listings_clean.drop(columns=['price']).values
y = listings_clean['price'].values

attrib_selection_model = ExtraTreesRegressor(n_estimators=100)
attrib_selection_model.fit(X,y)

# pontuação de importância para cada atributo (quanto maior a pontuação, mais importante é o atributo).
importances = attrib_selection_model.feature_importances_
importances_sorted_indexes = np.argsort(importances)[::-1]
print(f'Total de atributos: {len(importances_sorted_indexes)}\n')
for i, idx in enumerate(importances_sorted_indexes):
    print(f'{listings_clean.columns[idx]}: {importances[idx]}')
```

Decidimos então criar uma visão adicional dos dados onde descartamos atributos cuja importância não fosse equivalente a pelo menos 10% da importância do atributo mais relevante:

```python
# relevancia do atributo mais forte
ref_importance = importances[importances_sorted_indexes[0]]

# descartando os menores de 10%
importances_sorted_indexes_selected = []
for i, idx in enumerate(importances_sorted_indexes):
  if (importances[idx] > (ref_importance * 0.1)):
    importances_sorted_indexes_selected.append(idx)
  else:
    print(f'Descartando atributo: \'{listings_clean.columns[idx]}\'')

print(f'\nTotal de atributos descartados: {len(importances_sorted_indexes)-len(importances_sorted_indexes_selected)}\n')

# dataset com as colunas já excluídas
importances_sorted_indexes_selected.append(listings_clean.columns.get_loc('price')) # price
listings_clean_selected = listings_clean.iloc[:,importances_sorted_indexes_selected]
```

## Modelagem e Treinamento






### Modelagem


Finalizada a preparação dos dados, prosseguimos com o treinamento dos modelos.

Utilizamos a técnica de *pipelines*, que permite combinar de forma bastante prática as diferentes visões dos dados (original, padronizada e normalizada) com os difernetes modelos.

Levamos em conta um total de 5 modelos, que consideramos ser um conjunto razoavelmente representativo das diferentes técnicas candidatas para problemas de regressão:

1.   **LinearRegression**: regressão linear básica do tipo *least squares*.
2.   **KNeighborsRegressor**: k-vizinhos mais próximos.
3.   **DecisionTreeRegressor**: árvore de decisão.
4.   **ExtraTreesRegressor**:  *ensemble* do tipo *bagging*.
5.   **GradientBoosting**: *ensemble* do tipo *boosting*.

Definimos também o RMSE como medida de *scoring*, pois ela permite a interpretação do erro na escala do problema, o que é muito útil quando se trata de algo como *preço*.


A seguir a rotina de treinamento utilizada:

```python
# treina o modelo
def train_model(tag, kfold, X, y):

    # visoes dos dados
    data_views = [
          ('Original', None),
          ('Padronizada', StandardScaler()),
          ('Normalizada', MinMaxScaler())
    ]

    # modelos
    model_selection = [
      ('LinearRegression', LinearRegression()),
      ('KNeighborsRegressor', KNeighborsRegressor(n_neighbors=5)),
      ('DecisionTreeRegressor', DecisionTreeRegressor()),
      ('ExtraTreesRegressor', ExtraTreesRegressor(n_estimators=100, max_features='sqrt')),
      ('GradientBoostingRegressor', GradientBoostingRegressor(n_estimators=100, max_features='sqrt')),
    ]

    # definição do pipeline
    pipelines = []
    for model_label, model_operation in model_selection:
      for view_label, view_operation in data_views:
        pipeline_label =  model_label + '-' + tag + '-' + view_label
        pipeline_operation = Pipeline(steps=[(view_label,view_operation), (model_label,model_operation)])
        pipelines.append((pipeline_label, pipeline_operation))

    # resultados
    results = []
    names = []
    execution_times=[]

    # execução dos pipelines
    for name, model in pipelines:
      start_time = time.time()
      print(f'>>> {name} ...')
      cv_results = cross_val_score(model, X, y, cv=kfold, scoring='neg_root_mean_squared_error')
      elapsed_time_in_sec = time.time() - start_time
      execution_times.append(elapsed_time_in_sec)
      cv_results = -cv_results
      names.append(name)
      mean_result = cv_results.mean()
      print(f'\tMean RMSE: {mean_result:.3f}')
      results.append(mean_result)
      elapsed_time_str = ''.join(str(timedelta(seconds=(elapsed_time_in_sec))).split('.')[:1])
      print(f'\tTempo de Execução: {elapsed_time_str}')

    return results, names, execution_times
```

Realizamos então o treinamento utilizando as visões dos dados com todos os atributos (*Full*) e com alguns atributos descartados (*Selected*).

```python
# Visão SEM a seleção de atributos
results_full, names_full, execution_times_full = train_model('Full', kfold_obj, X_train_full, y_train_full)

# Visão COM a seleção de atributos
results_selected, names_selected, execution_times_selected = train_model('Selected', kfold_obj, X_train_selected, y_train_selected)
```

A seguir exibimos os resultados:

```python
# concatena os resultados
results = results_full + results_selected
names = names_full + names_selected
execution_times = execution_times_full + execution_times_selected

# ordena todas as listas simultaneamente de acordo com 'results'
results, names, execution_times = map(list, zip(*sorted(zip(results, names, execution_times))))
```

```python
# comparação dos modelos (RMSE)
fig = plt.figure(figsize=(25,5))
fig.suptitle('Comparação dos RMSE')
ax = fig.add_subplot(111)
ax.bar(names, results)
ax.set_xticklabels(names, rotation=90)
plt.ylabel('RMSE')
plt.show()

# comparação dos modelos (tempos de execução)
fig = plt.figure(figsize=(25,5))
fig.suptitle('Comparação dos Tempos de Execução')
ax = fig.add_subplot(111)
ax.bar(names, execution_times)
ax.set_xticklabels(names, rotation=90)
plt.ylabel('Tempo de Execução (s)')
plt.show()
```

### Otimização de Hiperparâmetros

Baseado nas observações realizadas durante a fase de modelagem, selecionamos o algoritmo *GradientBoostingRegressor* (mais preciso) para refinamento dos hiperparâmetros.

Pudemos verificar também que, dentro do *GradientBoostingRegressor*, a seleção de atributos não proporcionou diferença decisiva na precisão do algoritmo, portanto mantivemos somente o *dataset* completo (*Full*) para esta fase de otimização.

Escolhemos variar os hiperparâmetros *max_features* (controla o tamanho do subconjunto de atributos aletatórios usados nas árvores), *n_estimators* (o número de estágios de *boost*) e *learning_rate* (nível de contribuição de cada árvore). Estes dois últimos estão fortemente relacionados. Nos algoritmos de *gradient boosting* pode ocorrer *overfit* à medida que novas árvores são sequencialmente criadas e adicionadas ao modelo. O hiperparâmetro *learning_rate* é então importante para limitar a correção imposto pelas novas árvores a medida que são adicionadas.

```python
# selecionando o dataset
X = X_full
X_train = X_train_full
X_test = X_test_full
Y = y_full
y_train = y_train_full
y_test = y_test_full

# visoes dos dados
data_views = [
      ('Original', None),
      ('Padronizada', StandardScaler()),
      ('Normalizada', MinMaxScaler())
]

# modelos
model_selection = [
  ('GradientBoostingRegressor', GradientBoostingRegressor())
]

# hiperparametros
params_grid = [
  {
      'GradientBoostingRegressor__n_estimators':  [100, 300],
      'GradientBoostingRegressor__learning_rate': [0.1, 0.01],
      'GradientBoostingRegressor__max_features':  ['sqrt', 'auto'],
  }
]

# definicao do pipeline
pipelines = []
for model_idx, (model_label, model_operation) in enumerate(model_selection):
  for view_label, view_operation in data_views:
    pipeline_label =  model_label + '-' + view_label
    pipeline_operation = Pipeline(steps=[(view_label,view_operation), (model_label,model_operation)])
    pipelines.append((pipeline_label, pipeline_operation, params_grid[model_idx]))

# execução dos pipelines
best_score = 0
best_params = {}
best_name = ''

for name, model, param in pipelines:
  start_time = time.time()
  print(f'\n>>> {name} ...')
  grid = GridSearchCV(estimator=model, param_grid=param, scoring='neg_root_mean_squared_error', cv=kfold_obj, verbose=3)
  grid.fit(X_train, y_train)
  elapsed_time_in_sec = time.time() - start_time
  elapsed_time_str = (''.join(str(timedelta(seconds=(elapsed_time_in_sec))).split('.')[:1]))
  if best_score == 0 or -grid.best_score_ < best_score:
      best_score = -grid.best_score_
      best_params = grid.best_params_
      best_name = name

print(f'\nConfiguração mais precisa: \n\t{best_name} \n\tHiperparâmetros: {best_params} \n\tRMSE: {best_score}')
```

## Finalização do Modelo

O algoritmo *GradientBoostingRegressor*, com visão de dados padronizada, apresentou melhor precisão e portanto foi o modelo escolhido para finalização. O melhor conjunto de hiperparâmetros identificado foi *max_features=auto*, *learning_rate=0.1* e *n_estimators=300*.

```python
# padronização dos conjuntos de treino e teste
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# gera modelo
final_model = GradientBoostingRegressor(max_features='auto', learning_rate=0.1, n_estimators=300)
final_model.fit(X_train_scaled, y_train)

# aplicando o modelo ao conjunto de teste
y_predicted = final_model.predict(X_test_scaled)

print(f'Min y_test: {min(y_test)} \nMax y_test: {max(y_test)} \nMean y_test: {np.mean(y_test)}\n')
print(f'RMSE: {mean_squared_error(y_test, y_predicted)}')
```

```python
# número de predições a exibir
number_of_predictions_to_show = 50

pred_num = np.arange(number_of_predictions_to_show)
fig = plt.figure(figsize=(25,5))
fig.suptitle('Qualidade da Predição')
ax = fig.add_subplot(111)
ax.bar( pred_num - 0.15,  y_test[:number_of_predictions_to_show], width=0.3, label='Real')
ax.bar( pred_num + 0.15,  y_predicted[:number_of_predictions_to_show], width=0.3, label='Predito')
plt.xlabel('Teste')
plt.ylabel('Preço (log)')
plt.legend()
plt.show()
```
